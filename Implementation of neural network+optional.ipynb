```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Read train and test data
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')

# Checking data info
print("Train Data Info:")
print(train_data.info())
print("\nTest Data Info:")
print(test_data.info())

# Check for missing data
print("Missing values in each column:")
print(train_data.isnull().sum())
print(test_data.isnull().sum())

# Filling missing values of 'Age' and 'Fare' columns with mean
train_data['Age'].fillna(train_data['Age'].mean(), inplace=True)
test_data['Age'].fillna(test_data['Age'].mean(), inplace=True)

train_data['Fare'].fillna(train_data['Fare'].mean(), inplace=True)
test_data['Fare'].fillna(test_data['Fare'].mean(), inplace=True)

# Filling 'Embarked' column with the mode
train_data['Embarked'].fillna(train_data['Embarked'].mode()[0], inplace=True)
test_data['Embarked'].fillna(test_data['Embarked'].mode()[0], inplace=True)

# Convert categorical columns to numeric
train_data = pd.get_dummies(train_data, columns=['Sex', 'Embarked'], drop_first=True)
test_data = pd.get_dummies(test_data, columns=['Sex', 'Embarked'], drop_first=True)

# Normalization of 'Age' and 'Fare'
scaler = StandardScaler()
train_data[['Age', 'Fare']] = scaler.fit_transform(train_data[['Age', 'Fare']])
test_data[['Age', 'Fare']] = scaler.transform(test_data[['Age', 'Fare']])

# Select input properties and labels
Xtrain = train_data[['Pclass', 'Sex_male', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked_Q', 'Embarked_S']]
Ytrain = train_data['Survived']
Xtest = test_data[['Pclass', 'Sex_male', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked_Q', 'Embarked_S']]

Xtrain = Xtrain.replace({True: 1, False: 0})
Xtest = Xtest.replace({True: 1, False: 0})

# Convert to numpy arrays for neural network
Xtrain = np.array(Xtrain, dtype=float)
Xtest = np.array(Xtest, dtype=float)
Ytrain = np.array(Ytrain).reshape(-1, 1)  # Ensure Ytrain is reshaped to (m, 1)

# Neural network initialization function
def initialize_network(input_size, hidden_size, output_size):
    np.random.seed(42)
    W1 = np.random.randn(input_size, hidden_size) * 0.01  # Weights between input and hidden layer
    b1 = np.zeros((1, hidden_size))  # Biases of hidden layer
    W2 = np.random.randn(hidden_size, output_size) * 0.01  # Weights between hidden and output layer
    b2 = np.zeros((1, output_size))  # Biases of output layer
    return W1, b1, W2, b2

# Neural network functions
def forward_pass(X, W1, b1, W2, b2):
    z1 = np.dot(X, W1) + b1
    a1 = np.tanh(z1)
    z2 = np.dot(a1, W2) + b2
    y_hat = 1 / (1 + np.exp(-z2))  # Sigmoid output
    return y_hat, z1, a1

def backpropagation(X, y, y_hat, z1, a1, W2):
    m = X.shape[0]
    dz2 = (y_hat - y).reshape(m, 1)
    dW2 = np.dot(a1.T, dz2) / m
    db2 = np.sum(dz2, axis=0, keepdims=True) / m
    dz1 = np.dot(dz2, W2.T) * (1 - np.tanh(z1) ** 2)
    dW1 = np.dot(X.T, dz1) / m
    db1 = np.sum(dz1, axis=0, keepdims=True) / m
    return dW1, db1, dW2, db2

def update_parameters(W1, b1, W2, b2, dW1, db1, dW2, db2, learning_rate):
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2
    return W1, b1, W2, b2

def predict(X, W1, b1, W2, b2):
    y_hat, _, _ = forward_pass(X, W1, b1, W2, b2)
    predictions = (y_hat > 0.5).astype(int)
    return predictions

# Training and evaluation function
def train_and_evaluate(Xtrain, Ytrain, Xtest, hidden_size, num_iterations, learning_rate):
    input_size = Xtrain.shape[1]
    output_size = 1
    W1, b1, W2, b2 = initialize_network(input_size, hidden_size, output_size)
    losses = []

    for i in range(num_iterations):
        y_hat, z1, a1 = forward_pass(Xtrain, W1, b1, W2, b2)
        loss = np.mean(-Ytrain * np.log(y_hat) - (1 - Ytrain) * np.log(1 - y_hat))
        losses.append(loss)

        dW1, db1, dW2, db2 = backpropagation(Xtrain, Ytrain, y_hat, z1, a1, W2)
        W1, b1, W2, b2 = update_parameters(W1, b1, W2, b2, dW1, db1, dW2, db2, learning_rate)

        if i % 100 == 0:
            print(f"Iteration {i}, Loss: {loss:.4f}")

    predictions = predict(Xtest, W1, b1, W2, b2)
    return predictions, losses

# Visualizing loss curve
def plot_loss_curve(losses, learning_rate, hidden_size):
    plt.plot(losses)
    plt.title(f"Loss Curve (Learning rate={learning_rate}, Hidden size={hidden_size})")
    plt.xlabel("Iteration")
    plt.ylabel("Loss")
    plt.show()

# Trying different configurations
learning_rates = [0.001, 0.01, 0.1]  # Different learning rates to experiment
hidden_sizes = [4, 8, 16]  # Different number of neurons in hidden layer

for lr in learning_rates:
    for hidden_size in hidden_sizes:
        print(f"\nTraining with learning rate={lr} and hidden size={hidden_size}")
        predictions, losses = train_and_evaluate(Xtrain, Ytrain, Xtest, hidden_size, num_iterations=1000, learning_rate=lr)

        # Plotting the loss curve
        plot_loss_curve(losses, lr, hidden_size)

        # Save predictions to CSV file
        submission = pd.DataFrame({
            "PassengerId": test_data['PassengerId'],
            "Survived": predictions.flatten()
        })
        submission.to_csv(f'titanic_predictions_lr{lr}_hs{hidden_size}.csv', index=False)

        print(f"File saved as titanic_predictions_lr{lr}_hs{hidden_size}.csv")
```
